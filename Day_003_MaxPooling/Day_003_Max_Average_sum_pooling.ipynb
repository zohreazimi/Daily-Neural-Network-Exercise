```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

# لود دیتای MNIST
data = np.load(r"D:\GitHubProjects\Daily-Neural-Network-Exercise\data\mnist.npz")
x_train, y_train = data['x_train'], data['y_train']

# انتخاب یک تصویر
image = x_train[0]  # اولین تصویر
label = y_train[0]

# نرمال‌سازی و تبدیل به Tensor
image_tensor = torch.tensor(image, dtype=torch.float32).unsqueeze(0).unsqueeze(0)  
# شکل: (1, 1, 28, 28)

plt.imshow(image, cmap='gray')
plt.title(f"Label: {label}")
plt.show()
```

```python
# تعریف دو نوع pooling
max_pool = nn.MaxPool2d(kernel_size=2, stride=2)
avg_pool = nn.AvgPool2d(kernel_size=2, stride=2)

# اعمال روی تصویر
max_pooled = max_pool(image_tensor)
avg_pooled = avg_pool(image_tensor)

# نمایش نتایج
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(max_pooled.squeeze().numpy(), cmap='gray')
axes[0].set_title("Max Pooling")

axes[1].imshow(avg_pooled.squeeze().numpy(), cmap='gray')
axes[1].set_title("Average Pooling")

plt.show()
```

```python
kernel_size = 2
stride = 2

# AveragePooling
avg_pool = nn.AvgPool2d(kernel_size=kernel_size, stride=stride)

# SumPooling = AveragePooling × (تعداد پیکسل‌های هر پنجره)
sum_pooled = avg_pool(image_tensor) * (kernel_size * kernel_size)

# نمایش نتیجه
plt.imshow(sum_pooled.squeeze().numpy(), cmap='gray')
plt.title("Sum Pooling")
plt.show()
```
